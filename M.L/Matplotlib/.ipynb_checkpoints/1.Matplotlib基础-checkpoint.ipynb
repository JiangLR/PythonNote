一般只使用matplotlib模块下的pyplot子模块

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0, 10, 100)
```

```python
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y)
```

```python
cosy = np.cos(x)
```

```python
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = 'red')
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, linestyle = '--')
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.xlim(-5,15)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.ylim(0, 1.5)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.axis([-1, 11, -2, 2])
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
plt.plot(x, siny, label='sin(x)')
plt.plot(x, cosy, label='cos(x)')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

```python
plt.plot(x, siny, label='sin(x)')
plt.plot(x, cosy, label='cos(x)')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.title('Welcome to the ML. World')
```
